```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### read image

```python
img = cv2.imread("data/male0.jpg")
```

### Convert into GrayScale

```python
gray_pic = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Applying haar_cascade Classifiers

```python
haar = cv2.CascadeClassifier("data/haarcascade_frontalface_default.xml")
faces = haar.detectMultiScale(gray_pic,1.3,5)
```

```python
faces
```

```python
cv2.rectangle(img,(152,65),(152+151, 65+151),(0,0,255),5)
plt.imshow(img)
```

```python
cv2.imshow("Face_Detect",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Crop the Face

```python
crop_face = img[65:65+151, 152:152+151]
plt.imshow(crop_face)
```
