# Matplotlib Workflow
WE want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplot

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [5, 10, 15, 20]
y = [9, 18, 27, 36]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title ="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & Show (Saving the whole thing)

fig.savefig("images/sample-plot.png")

```

```python
# Bar plots
nut_butter_prices = {"Almond Butter":10,
                    "Peanut Butter":5,
                    "Cashew Butter":15}


fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());


ax.set(title="NutButter Store",
      ylabel="Price",
      xlabel="Products Names");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax.set(title="NutButter Store",
      ylabel="Products Names",
      xlabel="Price");
```
